```python
import csv
import pathlib
import pandas as pd
```

```python
regular_raw = pathlib.Path('../data/Regular_raw.csv')
regular_raw_df = pd.read_csv(regular_raw, index_col=False)
regular_raw_df.head()
```

```python
# Drop unused columns
regular_clean_df = regular_raw_df.drop(['Index',
                   'state',
                   'county',
                   'num_of_units',
                   'sqft',
                   'lot_size',
                   'days_on_market',
                   'year_built',
                   'tax',
                   'tax_history',
                   'videos',
                   'virtual_tours',
                   'parking_spots',
                   'parking_type',
                   'walkscore',
                   'mls_id',
                   'image',
                   'extra_images',
                   'property_type',
                   'property_sub_type',
                    'listhub_key',
                    'broker_name',
                    'broker_number',
                    'broker_url',
                    'source',
                    'mls_name',
                    'original_source',
                    'architecture_style',
                    'has_pool',
                    'is_water_front',
                    'heating_system',
                    'cooling_system',
                    'view_type',
                    'schools',
                    'parcel_number',
                    'url',
                    'disclaimer',
                    'listing_date',
                    'modification_timestamp',
                    'created_at',
                    'updated_at'                                        
                  ],axis=1)
```

```python
# rename and reorder columns
regular_clean_df['source'] = 'Regular'
regular_clean_df['propertyId'] = ''
regular_clean_df['nightPrice'] = ''
regular_clean_df['capacityOfPeople'] = ''
regular_clean_df['numOfRooms'] = ''
regular_clean_df['reviewsCount'] = ''
regular_clean_df['startRating'] = ''
regular_clean_df['occupancy'] = ''
regular_clean_df['nightsBooked'] = ''
regular_clean_df['clean_mthly_price'] = ''
regular_clean_reorder_df = regular_clean_df[['id',
                                  'propertyId',
                                  'source',
                                  'neighborhood_id',
                                  'description',
                                  'title',
                                  'nightPrice',
                                  'baths',
                                  'beds',
                                  'price',
                                  'address',
                                  'neighborhood',
                                  'city',
                                  'zipcode',
                                  'type',
                                  'lat',
                                  'lon',
                                  'capacityOfPeople',
                                  'numOfRooms',
                                  'reviewsCount',
                                  'startRating',
                                  'occupancy',
                                  'nightsBooked',
                                  'clean_mthly_price']]
regular_clean_reorder_df.head()
```

```python
regular_clean_reorder_df.rename({'title': 'Title',
                                     'baths': 'numOfBaths',
                                     'beds': 'numOfBeds',
                                     'zipcode': 'zip',
                                     'price': 'Price',
                                     'type': 'propertyType'},axis = 1, inplace = True)
```

```python
regular_clean_reorder_df.head()
```

```python
regular_clean_reorder_df.to_csv('../data/Regular_clean_v1.csv', index = False)
```
