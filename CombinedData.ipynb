```python
import pandas as pd
import csv
import pathlib

csvpath = pathlib.Path('Airbnb_raw - Sheet1.csv')

data_airbnb_df = pd.read_csv(csvpath)
```

```python
# add column for Source Type
airbnb_occupancy = 18
source_Airbnb = "Airbnb"
data_airbnb_df["source"] = source_Airbnb # added source column
data_airbnb_df["price"] = data_airbnb_df["nightPrice"]*airbnb_occupancy # added price column by multiplying nightPrice column by a constant of 25
data_airbnb_df.rename(columns={'airbnbNeighborhood': 'neighboorhood', 'propertyType': 'type'}, inplace=True)
clean_airbnb_data = data_airbnb_df[["id", "neighborhood", "price", "type", "lat", "lon", "source"]]
clean_airbnb_data.set_index("id", inplace=True)
clean_airbnb_data.head()
```

```python
air_neighborhoods = clean_airbnb_data.groupby("neighborhood").price.count().sort_values(ascending=False)
air_neighborhoods.head(40)
```
