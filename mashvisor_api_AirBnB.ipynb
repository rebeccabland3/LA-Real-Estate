```python
import requests
from config import api_key
```

```python
import pandas as pd
```

```python
headers = {
    'x-api-key': api_key
}
```

## Get Los Angeles Neighborhoods


```python
state = 'CA'
city = 'Los%20Angeles'
```

```python
url =  f'https://api.mashvisor.com/v1.1/client/city/neighborhoods/{state}/{city}'
```

```python
response = requests.get(url, headers=headers)
```

```python
results_list = response.json()['content']['results']
results_list
```

```python
los_angeles_neighborhoods_df = pd.DataFrame(results_list)
los_angeles_neighborhoods_df
```

## Get Airbnb Info for Neighborhoods

```python
all_neighborhoods_list = []
```

```python
#neighborhood_id = 7877
for neighborhood_id in los_angeles_neighborhoods_df['id']:
    airbnb_url = f'https://api.mashvisor.com/v1.1/client/neighborhood/{neighborhood_id}/airbnb/details'
    params = {
        'state': 'CA',  # required per API documentation
        'items': 20     # appears to max at 20
    }
    try:
        response = requests.get(airbnb_url, params=params, headers=headers)
    except Exception as e:
        f"Failed to get response: {e}"

    try:
        properties = response.json()['content']['properties']
    except Exception as e:
        f"Failed to extract properties from response: {e}"
    else:
        all_neighborhoods_list = all_neighborhoods_list + properties
        
```

```python
all_neighborhoods_df = pd.DataFrame(all_neighborhoods_list)
all_neighborhoods_df
```

```python
all_neighborhoods_df.to_csv('../all_neighborhoods.csv')
```
