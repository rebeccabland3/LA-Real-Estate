```python
import csv
import pathlib
import pandas as pd
```

```python
# Read in clean Regular data
Regular_clean = pathlib.Path('../data/Regular_clean_v1.csv')
Regular_clean_df = pd.read_csv(Regular_clean, index_col=False)
```

```python
# Read in raw Airbnb data
airbnb_raw = pathlib.Path('../data/Airbnb_raw.csv')
airbnb_raw_df = pd.read_csv(airbnb_raw, index_col=False)
```

```python
# Drop unused columns
airbnb_clean_df = airbnb_raw_df.drop(['Index',
                   'status',
                   'nightPriceÙ€native',
                   'weeklyPrice',
                   'monthlyPrice',
                   'cleaningFeeNative',
                   'image',
                   'rentalIncome',
                   'state',
                   'roomType',
                   'roomTypeCategory',
                   'amenities',
                   'reviews',
                   'createdAt',
                   'updatedAt',
                   'url',
                   'rental_income',
                   'neighborhood',
                   'nightRate',
                   'airbnbZIP'
                   
                  ],axis=1)
```

```python
# rename and reorder columns
airbnb_clean_df['Title'] = ''
airbnb_clean_df['Price'] = ''
airbnb_clean_df['clean_mthly_price'] = ''
airbnb_clean_reorder_df= airbnb_clean_df[['id',
                                  'propertyId',
                                  'source',
                                  'airbnbNeighborhoodId',
                                  'name',
                                  'Title',
                                  'nightPrice',
                                  'numOfBaths',
                                  'numOfBeds',
                                  'Price',
                                  'address',
                                  'airbnbNeighborhood', 
                                  'airbnbCity',
                                  'zip',
                                  'propertyType',
                                  'lat',
                                  'lon',
                                  'capacityOfPeople',
                                  'numOfRooms',
                                  'reviewsCount',
                                  'startRating',
                                  'occupancy',
                                  'nightsBooked',
                                  'clean_mthly_price']]
airbnb_clean_reorder_df.head()
```

```python
# rename columns so both Regular and Airbnb data are aligned
airbnb_clean_reorder_df.rename({'airbnbNeighborhoodId': 'neighborhood_id',
                                     'name': 'description',
                                     'airbnbCity': 'city',
                                     'airbnbNeighborhood': 'neighborhood'},axis = 1, inplace = True)
```

```python
airbnb_clean_reorder_df.head()
```

```python
# merge the two datasets together
air_reg_merged_df = pd.concat([airbnb_clean_reorder_df, Regular_clean_df])
```

```python
# export combined datasets to a csv
air_reg_merged_df.to_csv('../data/combined_clean_v1.csv',index = False)
```
