```python
# Dependencies
import pandas as pd
```

```python
# Store filepath in a variable
#Source: kaggle - 2016 FBI reports
crime_csv = "ca_offenses_by_city2.csv" ####### wrong csv
```

```python
# Read Crime Data file with the pandas library
crime_df = pd.read_csv(crime_csv)
```

```python
# Show just the header
crime_df.head()
```

```python
clean_df = crime_df.drop(columns= "Rape (legacy definition)")
clean_df
```

```python
clean_df.describe()
```

```python
clean_df = clean_df.rename(columns={ "City": "neighborhood name"})
clean_df.head()
```

```python
#Calculated Crime rate [CR Formula = (# reported crime/total population)*100,000] ~interpretation: per 100,000 of general population, there are 1,549.83 crimes that occur in Agoura Hills
clean_df["Total Crimes"] = clean_df["Violent crime"] + clean_df["Property crime"]
clean_df["Crime Rate"]= round((clean_df["Total Crimes"]/clean_df["Population"])*100000, 2)
clean_df.head()
```

```python
#Use to look up neighborhoods that will match clean data (for the purposes of inner join matching)
unique_neigbh = clean_df["neighborhood name"].unique()
unique_neigbh
```

```python
len(unique_neigbh)
```

```python
crime_rate_df = clean_df[["neighborhood name", "Crime Rate"]]
crime_rate_df.head(40)
```

```python
# Export file as a CSV, without the Pandas index, but with the header (file to be used in merging)
clean_df.to_csv("Output/crime_rate.csv", index=False, header=True)
```
